## Função de gerar o boxplot


```python
def gera_boxplot(dataset,titulo,subtitulo):
  sns.set_theme(palette='pastel',font_scale=1.2)
  plt.figure(figsize=(10,8))

  ax = sns.boxplot(data = dataset)
  ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))

  plt.ylim(0,max(dataset.max())*1.05)
  plt.xlabel(None)
  plt.ylabel(None)

  ax.set_title(titulo+'\n',size=15,loc='center',weight='bold')
  ax.text(s=subtitulo,x=-0.5,y=max(dataset.max())*1.07,fontsize=13,ha='left')
```

## Distribuição dos empréstimos mensais pelos Alunos de Graduação

1. Há uma variação considerável dos empréstimos por mês (percebido pelo tamanho dos quartis)
2. Observando especialmente a mediana de cada ano, tem-se o mesmo comportamento do gráfico de linhas feito dos anos.
3. Percebemos que em alguns anos, houveram meses outliers em relação ao todo (2011, 2014, 2017 e 2018). Se fizermos a remoção destes valores, teremos médias maiores do que as apresentadas no gráfico.
4. Deve ser analisado o que ocorreu nos meses outliers, para que se possa saber o motivo do comportamento fora do padrão em relação aos demais meses.
5. Nos anos de 2014 e 2017, excluindo os outliers, temos uma distribuição mais uniforme, com menor desvio mensal de empréstimos.

# Estilizando a tabela e exportando HTML

```python
# Configurando o indice, e os nomes dos cursos.

tab_dif.reset_index(inplace=True)
tab_dif.columns = ['Curso','2017-2018','2018-2019','2019-2022']
tab_dif.Curso = tab_dif.Curso.str.capitalize()

tab_dif
```

```python
# Definindo os estilos para cabeçalho e texto conforme solicitado

th_props = [
    ('font-size','1.4rem'),
    ('text-align','center'),
    ('font-weight','bold'),
    ('color','whitesmoke'),
    ('background-color','#001692'),
    ('border-radius','0.25rem'),
    ('box-shadow','0 0 1rem gray')
]

td_props = [
    ('font-size','1rem'),
    ('padding','0.5rem'),
    ('text-align','left'),
    ('font-weight','bold'),
    ('border-bottom','0.1rem solid lightgray')
]

table_style = [
    dict(selector='th', props=th_props),
    dict(selector='td', props=td_props)
]
```

```python
tab_dif.style.hide(axis=0)\
              .format('{:.2f}%',subset=['2017-2018','2018-2019','2019-2022'])\
              .text_gradient(cmap='RdYlGn',subset=['2017-2018','2018-2019','2019-2022'],vmin=0,vmax=0.01)\
              .set_table_styles(table_style)
```

```python
tab_dif.style.hide(axis=0)\
              .format('{:.2f}%',subset=['2017-2018','2018-2019','2019-2022'])\
              .text_gradient(cmap='RdYlGn',subset=['2017-2018','2018-2019','2019-2022'],vmin=0,vmax=0.01)\
              .set_table_styles(table_style)\
              .to_html('tabela.html')
```
